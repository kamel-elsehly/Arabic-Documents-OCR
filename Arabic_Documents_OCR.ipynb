```python
# import library
import easyocr
#specify shortform of language you want to extract,
# I am using Hindi(hi) and English(en) here by list of language ids
reader = easyocr.Reader(['ar','en'])
```

```python
# Doing OCR. Get bounding boxes.
bounds = reader.readtext('/content/doc.jpg', detail=0) #detail=0 argument will only give text in array
print("Output:")
print(bounds)
```

```python
# Read Image
import PIL
im = PIL.Image.open("/content/doc.jpg") #provide your image path
im
```

```python
from PIL import Image, ImageDraw, ImageFont
# Draw bounding boxes
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds: # iterate though all the tuples of output
        p0, p1, p2, p3 = bound[0] # get coordinates 
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```
